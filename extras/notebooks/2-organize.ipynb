```python
import pandas as pd
import os
```

```python
path_images = '../../dev/imgs/temp'
path_res = '../../dev/images_new.csv'
path_animals = '../../dev/images.csv'
```

```python
df_animals = pd.read_csv(path_animals, sep=',')
df_animals['filename'] = 'media/' + df_animals['filename']
df_animals.head(2)
```

```python
animals = list(df_animals.columns[1:])
animals[:3]
```

```python
# list dirs in path_images_google
dirs = [ name for name in os.listdir(path_images) if os.path.isdir(os.path.join(path_images, name)) ]
res_table = []
for dirname in dirs:
    animal_dir = os.path.join(path_images, dirname)
    files = [f for f in os.listdir(animal_dir) if os.path.isfile(os.path.join(animal_dir, f))]
    animal_dir_2 = os.path.join(path_images, dirname + " animal")
    files_2 = [f for f in os.listdir(animal_dir) if os.path.isfile(os.path.join(animal_dir, f))]
    files = files + files_2
    for filename in files:
        file_path = os.path.join(animal_dir, filename)
        res_row = []
        path_images_relative= '../../'
        filename_ = file_path[6:]
        if ',' in filename_ or 'gif' in filename_:
            continue
        res_row.append(filename_)
        found = False
        for animal_name in animals:
            if animal_name in dirname:
                found = True
                res_row.append("1")
            else:
                res_row.append("0")
        if not found:
            print("animal " + animal_name + " not found")
        res_table.append(res_row)
#list files in each dir
# copy the file and append the row to the dataframe
# save the dataframe
```

```python
with open(path_res, 'w') as f:
    for _list in res_table:
        f.write(",".join(_list) + "\n")
```

```python
df_res = pd.read_csv(path_res, sep=',', header=None)
df_res.columns = df_animals.columns
df_res.head(2)
```

```python
dfs = []
dfs.append(df_animals)
dfs.append(df_res)
df = pd.concat(dfs)
```
